```python
#%pip install seaborn pandas matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загрузка данных
data = pd.read_csv('IRIS.csv')

# Просмотр первых строк датасета
data.head()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(data, hue='species', palette='Set2')  # hue указывает на признак, по которому происходит разделение
plt.title('Парная диаграмма ирисов', x=-1, y=4.3)
plt.show()
```

```python
plt.figure(figsize=(12, 10))

# Создание отдельных скрипичных диаграмм для каждой из четырех характеристик
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='species', y=feature, data=data, palette="Set2")
    plt.title(f'Скрипичная диаграмма для {feature}', fontsize=14)

plt.tight_layout()
plt.show()
```
